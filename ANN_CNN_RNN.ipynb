**ANN BACKPROPAGATION**

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Sample dataset
X = torch.randn(100, 10)   # 100 samples, 10 features
y = torch.randint(0, 2, (100, 1)).float()

# ANN model
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.fc1 = nn.Linear(10, 16)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.sigmoid(self.fc2(x))
        return x

model = ANN()

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Training loop
for epoch in range(50):

    outputs = model(X)
    loss = criterion(outputs, y)

    optimizer.zero_grad()
    loss.backward()      # Backpropagation
    optimizer.step()

    if epoch % 10 == 0:
        print("Epoch:", epoch, "Loss:", loss.item())
```

**CNN BACKPROPAGATION**

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Fake image data (batch, channel, height, width)
X = torch.randn(32, 1, 28, 28)
y = torch.randint(0, 10, (32,))

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv1 = nn.Conv2d(1, 8, kernel_size=3)
        self.pool = nn.MaxPool2d(2,2)
        self.fc1 = nn.Linear(8*13*13, 10)

    def forward(self, x):

        x = self.pool(torch.relu(self.conv1(x)))
        x = x.view(x.size(0), -1)
        x = self.fc1(x)

        return x

model = CNN()

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

for epoch in range(20):

    outputs = model(X)
    loss = criterion(outputs, y)

    optimizer.zero_grad()
    loss.backward()      # CNN backpropagation
    optimizer.step()

    print("Epoch:", epoch, "Loss:", loss.item())
```

**RNN BACKPROPAGATION**

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Sequence data (batch, sequence_length, features)
X = torch.randn(32, 5, 10)
y = torch.randint(0, 2, (32,1)).float()

class RNNModel(nn.Module):
    def __init__(self):
        super(RNNModel, self).__init__()

        self.rnn = nn.RNN(input_size=10, hidden_size=20, batch_first=True)
        self.fc = nn.Linear(20,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):

        out, hidden = self.rnn(x)
        out = out[:, -1, :]   # last timestep
        out = self.sigmoid(self.fc(out))

        return out

model = RNNModel()

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(30):

    outputs = model(X)
    loss = criterion(outputs, y)

    optimizer.zero_grad()
    loss.backward()      # Backpropagation Through Time (BPTT)
    optimizer.step()

    print("Epoch:", epoch, "Loss:", loss.item())
```
